<h1 style="color:Pink;">Introduction</h1>

<p style="font-size:20px;">
This project is going to leverage <span style="color:Pink;">ARIMA</span> models for forscasting Monthly Milk productions. To beign, let's understand what <span style="color:Pink;">ARIMA</span> models are and what they entailed.
</p>

<h4><span style="color:Pink;">ARIMA</span> -: <span style="color:Pink;">A</span>utoRegressive <span style="color:Pink;">I</span>ntegrated <span style="color:Pink;">M</span>oving Average :-</h4>

<p style="font-size:20px;">They are statistical technique for forcasting timeseries Data. They are a combination of three parts.</p>

<ul style="font-size:20px;">
<li> <span style="color:Pink;">AutoRegressive</span> :- Uses past data to forcast future values </li>
<li> <span style="color:Pink;">Integrated</span> :- Accounts for presence of non-stationarity in Time Series Data </li>
<li> <span style="color:Pink;">Moving Average</span> :- Uses Error from the past forcast to improve future forcast </li>
</ul>

<p style="font-size: 20px;">ARIMA models are denoted as <span style=" background-color: #383E4A; color:#D1B56D; border-radius: .375rem; padding: 0px 5px; font-weight:400px;" > ARIMA(p, d, q)</span></p>

<h4> <span style="color:#D1B56D; font-weight: bold; background-color:#383E4A; padding: 0px 5px; border-radius: .375rem">p</span>:- number of autoregressive terms </h4>

<p style="font-size: 20px;">say <span style=" background-color: #383E4A; color:#D1B56D; border-radius: .375rem; padding: 0px 5px; font-weight:400px;" > ARIMA(2, 0, 0)</span> <br> <br> This model has two AutoRegressive Terms ( <em>it use two immediate past information to predict the current values)</em> </p>

$  X_t = c + \phi_1 \cdot X_{t-1} + \phi_2 \cdot X_{t-2} + \epsilon_t $

$
\left\{
\substack{
    \text{first immediate data } X_{t-1} \text{ and weight } \phi_1 \\
    \text{second immediate data } X_{t-2} \text{ and weight } \phi_2
}
\right\}
$

<p style="font-size: 20px;">The model Combine this two past values to forcast curret values</p>

<h4> <span style="color:#D1B56D; font-weight: bold; background-color:#383E4A; padding: 0px 5px; border-radius: .375rem">d</span>:- number of integrated terms </h4>

<p style="font-size: 20px;">say <span style=" background-color: #383E4A; color:#D1B56D; border-radius: .375rem; padding: 0px 5px; font-weight:400px;" > ARIMA(2, 3, 0)</span> <br> <br> This model has three Integrated Terms ( <em>number of differencing that has to be made to make the data stationary)</em> </p>

<p style="font-size: 20px;"><span style="color:Pink;">Differencing</span> is subtracting each observation from it's previous observations to make the series stationary.</p>

<p style="font-size: 20px;"><span style="color:Pink;">Stationary</span> time seires is series with :- </p>



<ul style=font-size:20px;">
<li> constant mean</li>
<li> constant standard Deviations </li>
<li> no seasonality </li>
</ul>

<p style="font-size: 20px;">for 3 differencing </p>
<p style="font-size: 20px;"><span style="color:#D1B56D; font-weight: bold; background-color:#383E4A; padding: 0px 5px; border-radius: .375rem">1st Differencing</span> fd </p>

$fd$ -->  $ X_t - X_{t-1} $

<p style="font-size: 20px;"><span style="color:#D1B56D; font-weight: bold; background-color:#383E4A; padding: 0px 5px; border-radius: .375rem">2nd Differencing</span> sd </p>

$sd$ --> $ fd_t - fd_{t-1} $


<p style="font-size: 20px;"><span style="color:#D1B56D; font-weight: bold; background-color:#383E4A; padding: 0px 5px; border-radius: .375rem">3rd Differencing</span> td </p>

$td$ --> $ sd_t - sd_{t-1} $







<p style="font-size: 20px;">The diferencing may continue until the timeseries data is stationary (<em> other differencing   `seasonal difference` `first seasonal difference` etc</em>)</p>


<h4> <span style="color:#D1B56D; font-weight: bold; background-color:#383E4A; padding: 0px 5px; border-radius: .375rem">q</span>:- number of Moving Average Terms </h4>

<p style="font-size: 20px;">say <span style=" background-color: #383E4A; color:#D1B56D; border-radius: .375rem; padding: 0px 5px; font-weight:400px;" > ARIMA(2, 3, 1)</span> <br> <br> This model has 1 Moving Average Terms ( <em>number of lagged error (residual) from the previous time was used to forcast the next value)</em> </p>


$ X_t = ... + \epsilon_t + \theta * \epsilon_{t-1} $

$ X_t $ the current time series value 

$ \epsilon_t $ current error term $ \theta $ error term weight

$ \epsilon_{t-1} $ one previous error term 

<h4><span style="color:Pink;">ARIMA</span> -: are used for :-</h4>

<h4><span style="color:Pink;">ACF</span> -: <span style="color:Pink;">A</span>uto<span style="color:Pink;">C</span>orrelation <span style="color:Pink;">F</span>unction :-</h4>

<p style="font-size:20px;">This function plots the correlations of a time series and the  lagged version of itself</p>

<p style="font-size: 20px;">say <span style="color:#D1B56D; font-weight: bold; background-color:#383E4A; padding: 0px 5px; border-radius: .375rem">Jan</span>, <span style="color:#D1B56D; font-weight: bold; background-color:#383E4A; padding: 0px 5px; border-radius: .375rem">Feb</span>,  <span style="color:#D1B56D; font-weight: bold; background-color:#383E4A; padding: 0px 5px; border-radius: .375rem">Mar</span>, <span style="color:#D1B56D; font-weight: bold; background-color:#383E4A; padding: 0px 5px; border-radius: .375rem">Apr</span> time series data is obtained, <span style="color:Pink;">ACF</span> will be used to  plot the correlation (<em style="color:Tan">spearaman rank or pearson </em> ) between <span style="color:#D1B56D; font-weight: bold; background-color:#383E4A; padding: 0px 5px; border-radius: .375rem">Jan</span> (lag) vs <span style="color:#D1B56D; font-weight: bold; background-color:#383E4A; padding: 0px 5px; border-radius: .375rem">Feb</span> (future), <span style="color:#D1B56D; font-weight: bold; background-color:#383E4A; padding: 0px 5px; border-radius: .375rem">Feb</span> (lag) vs <span style="color:#D1B56D; font-weight: bold; background-color:#383E4A; padding: 0px 5px; border-radius: .375rem">Mar</span> (future), <span style="color:#D1B56D; font-weight: bold; background-color:#383E4A; padding: 0px 5px; border-radius: .375rem">Mar</span> (lag) vs <span style="color:#D1B56D; font-weight: bold; background-color:#383E4A; padding: 0px 5px; border-radius: .375rem">Apr</span> (future) this represent a lag of one month (it can be month, year etc)</p>



<h4><span style="color:Pink;">PACF</span> -: <span style="color:Pink;">P</span>artial <span style="color:Pink;">A</span>uto<span style="color:Pink;">C</span>orrelation <span style="color:Pink;">F</span>unction :-</h4>

<p style="font-size:20px;">This function plots the correlations of a time series and the  lagged version of itself it controls the effect of the lag in between future and prefered time period lag</p>





<p style="font-size: 20px;">say we looking at the correlation between current period and 3 month lag period. PACF will  calculate the correlation and also account for or control the effects of the 1 month and 2 month lag in between the current period and 3 month lag period  PACF uses regression model to achieve this.</p>

$$ X_t = \beta_0 + \beta_1 \cdot X_{t-1} + \beta_2 \cdot X_{t-2} +  \beta_3 \cdot X_{t-3} $$

$ X_t $ is the current period

$ X_{t-3} $ is the lag period 3

<p style="font-size: 20px;">1 and 2 lag period will be controlled</p>

Random Search and Grid Search are some techniques to be used in Search for the optimal parameter   AIC (<em>Akaike Information Criteria</em>) and BIC(<em>Bayesian Information Criteria</em>) are functions that are used to meseaure the performance on the validation set</p>

```python
from statsmodels.tsa.stattools import adfuller
```

```python
result = adfuller(df['Monthly milk production (pounds per cow)'])
```

```python
result
```

### The result of this test provides some statistics which can be formatted with a heuristic function that tells if the null is rejected or failed to be rejected

```python
def adf_checker(timeseries):
    print("Augmented Dickey FUller Test")
    results = adfuller(df['Monthly milk production (pounds per cow)'])
    labels = ["ADF Test Statistic", "p-value", "No. of lags", "No of observation"]
    for value, label in zip(results, labels):
        print(f"{label} : {value}")

    if results[1] <= 0.05:
        print("Strong evidence against the null hypothesis")
        print("Reject the null hypothesis")
        print("Time series has no unit root hence stationary")
    else:
        print("Weak evidence against the null hypothesis")
        print("Failed to reject the null hypothesis")
        print("Time series has unit root hence non-stationary")
```

### We can check the stationarity of te data

```python
adf_checker(df['Monthly milk production (pounds per cow)'])
```

<p style="font-size: 20px;">At times finding the  <span style="color:#D1B56D; background-color:#383E4A; padding:5px; border-radius:5px">p</span> and <span style="color:#D1B56D; background-color:#383E4A; padding:5px; border-radius:5px">q</span> values from <span style="color:Pink; background-color:#383E4A; padding:5px; border-radius:5px">ACF</span>  and  <span style="color:Pink; background-color:#383E4A; padding:5px; border-radius:5px">PACF</span> plot may seem difficult and requires experiences. Other methods exits such pick small values and iterating over different values and evaluating each pair of value with <span style="color:Pink; background-color:#383E4A; padding:5px; border-radius:5px">AIC & BIC</span></p>

<p style="font-size: 20px;"> From the first difference ACF and PACF plot <span style="color:#D1B56D; background-color:#383E4A; padding:5px; border-radius:5px">p</span> will be <em>16</em> and <span style="color:#D1B56D; background-color:#383E4A; padding:5px; border-radius:5px">q</span> will be <em>18</em>. This values are possible but are too high and will make our model complex which may lead to overfitting. In that case we can use other statistical means to obtain <span style="color:#D1B56D; background-color:#383E4A; padding:5px; border-radius:5px">p</span> and <span style="color:#D1B56D; background-color:#383E4A; padding:5px; border-radius:5px">q</span></p>

<p style="font-size: 20px;">Using AIC (<em>  Akaike Information Criteria </em>)  to determine the optimal <span style="color:#D1B56D; background-color:#383E4A; padding:5px; border-radius:5px">p</span> and <span style="color:#D1B56D; background-color:#383E4A; padding:5px; border-radius:5px">q</span> values</p>

```python
from statsmodels.tsa.arima.model import ARIMA

def aic_optimal_pq(timeseries, max_param):

    best_aic = np.inf
    best_order = None

    for p in range(max_param):
        for q in range(max_param):

            model= ARIMA(timeseries, order=(p, 0, q), freq="MS")
            result = model.fit()
            current_aic = result.aic

            if current_aic < best_aic:
                best_aic = current_aic
                best_order = (p, 0, q)

    return {'best aic' : current_aic, "best order":best_order}
```

```python
aic_optimal_pq(df['First Difference'], 3)
```

<p style="font-size: 20px;">Using BIC (<em>  Bayesian Information Criteria </em>)  to determine the optimal <span style="color:#D1B56D; background-color:#383E4A; padding:5px; border-radius:5px">p</span> and <span style="color:#D1B56D; background-color:#383E4A; padding:5px; border-radius:5px">q</span> values</p>

```python
def bic_optimal_pq(timeseries, max_param):

    best_bic = np.inf
    best_order = None

    for p in range(max_param):
        for q in range(max_param):

            model= ARIMA(timeseries, order=(p, 0, q), freq="MS")
            result = model.fit()
            current_bic = result.bic

            if current_bic < best_bic:
                best_aic = current_bic
                best_order = (p, 0, q)

    return {'best aic' : current_bic, "best order":best_order}
```

```python
bic_optimal_pq(df['First Difference'], 3)
```

<p style="font-size: 20px;">From the heuristic function <span style="color:pink; background-color:#383E4A; padding:5px; border-radius:5px">aic_optimal_pq</span> and <span style="color:pink; background-color:#383E4A; padding:5px; border-radius:5px">bic_optimal_pq</span> for (<em>  Akaike Information Criteria </em> and <em>Bayesian Information Criteria </em>)  the best <span style="color:#D1B56D; background-color:#383E4A; padding:5px; border-radius:5px">p</span> and <span style="color:#D1B56D; background-color:#383E4A; padding:5px; border-radius:5px">q</span> values are 2 for both</p>

<p style="font-size: 20px;">creating the model now use the optimal  <span style="color:#D1B56D; background-color:#383E4A; padding:5px; border-radius:5px">p</span> and <span style="color:#D1B56D; background-color:#383E4A; padding:5px; border-radius:5px">q</span> values obtain</p>

```python
model = ARIMA(df['First Difference'], order=(2, 0, 2), freq="MS", seasonal_order=(1, 1, 1, 12))
result = model.fit()
```

<p style="font-size: 20px;">The model can be interpreted as using 2 <span style="color:#D1B56D; background-color:#383E4A; padding:5px; border-radius:5px">AR(2)</span> and  0 <span style="color:#D1B56D; background-color:#383E4A; padding:5px; border-radius:5px">I(0)</span> since the data was made stationary from the first difference and 2 lag errors <span style="color:#D1B56D; background-color:#383E4A; padding:5px; border-radius:5px">MA(2)</span> to predict the current value</p>
